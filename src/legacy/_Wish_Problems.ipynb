public class Solution {
    private class Edge{
        String name;
        String relation;
        public Edge (){}
        public Edge(String name, String relation) {
            this.name = name;
            this.relation = relation;
        }
    }
    private Map<String, List<Edge>> graph;
    private Deque<Edge> path;
    private Set<Edge> marked;
    private String source;
    private String target;
    public void relationship(List<List<String>> relations ,String source, String target) {
        this.graph = new HashMap<>();
        buildGraph(relations);
        this.path = new ArrayDeque<>();
        this.marked = new HashSet<>();
        this.source = source;
        this.target = target;
        dfs(source);
    }
    public void dfs(String cur){
        if( cur.equals(target) ) {
            printPath(path);
        }
        if( !graph.containsKey(cur) ) {
            return;
        }
        for(Edge edge : graph.get(cur)){
            if( marked.contains(edge) ) {
                continue;
            }
            marked.add(edge);
            path.addLast(edge);
            dfs(edge.name);
            marked.remove(edge);
            path.removeLast();
        }
        return;
    }
    public void printPath(Deque<Edge> path){
        StringBuilder sb = new StringBuilder();
        sb.append(source);
        
        for(Edge edge: path){
            sb.append(" ");
            sb.append(edge.relation);
            sb.append(" ");
            sb.append(edge.name);
        }
        System.out.println(sb.toString());
    }
    
    public void buildGraph( List<List<String>> relations ) {
        for( List<String> relation : relations ) {
            String start = relation.get(0);
            Edge edge = new Edge(relation.get(2), relation.get(1));
            graph.putIfAbsent(start, new ArrayList<Edge>());
            graph.get(start).add(edge);
        }
    }

        Solution solution = new Solution();

        l.add(Arrays.asList("C", "loves", "D"));
        l.add(Arrays.asList("D", "loves", "B"));

public class Main {
    public static int waysSumToTarget(int[][] board, int target){
        int m = board.length;
        int n = board[0].length;
        int count = 0;
        ArrayDeque<int[]> dfs = new ArrayDeque<>();
        dfs.offer(new int[]{0,0, board[0][0]});
        while( !dfs.isEmpty() ){
            int[] cur = dfs.poll();
            int row = cur[0];
            int col = cur[1];
            int sum = cur[2];
            if( row == m - 1 && col == n - 1){
                if(sum == target){
                    count += 1;
                }
                continue;
            }
            if(row + 1 < m ){
                dfs.offer(new int[]{row + 1, col, sum + board[row + 1][col] });
            }
            if(col + 1 < n){
                dfs.offer(new int[]{row, col + 1, sum + board[row][col + 1] });
            }
        }
        return count;
    }

#### HashMap 
[参考了这个](https://yikun.github.io/2015/04/01/Java-HashMap%E5%B7%A5%E4%BD%9C%E5%8E%9F%E7%90%86%E5%8F%8A%E5%AE%9E%E7%8E%B0/)

* General :       
    HashMap is a LinkedList Array. It stores a key value pair as a node.     
    Stores no order and order may change.
* Field :      
    Capacity = 8, loadFactor 0.75.    
    If more than this propotion of buckets are taken, resize needed.    
* Node :     
    Data stored as Nodes.     
    Nodes has fields of key, value, next and hash.   
* Calculate hash from hashcode():     
    hashCode() returns an Integer, which is 32 digits, and we do a bitwise XOR between first 16 digits and last 16 digits.     
* Calculate index from node's hash:   
    Do a bitwise AND between the capacity - 1 and node's hash value.           
    (n - 1) & hash     
    since n = 2^k, n - 1 = 2^k - 1 = 111...11 (total k digits),     
    hash & (n - 1) = take the last k digit of hash = hash % n;     
* Collision:   
    if collision, store key,value pairs as LinkedList.  
    if key exist, update value.      
    if collision too much, linkedlist to long, linkedlist turns into red-black tree.  This is after Java8.    
* Resize:
    We double size. We don't need to re-calculate hash of each node, just bitwise AND again. 
* get method:

#### Heap 
How to implement min heap take Integer type for example. 
* General:     
    Heap is a complete binary tree. Or has the structure of complete binary tree.     
    The value of each node is smaller or equal than the value of its children's.   
    Complete means it won't goes to next level until current tree is full. 
* Class field:      
   int[] heap, int size.   
   An array and size, size is important.        
* Method push/put:          
   We push new element to the index = size. i.e we put it to the end of the array.  
   And then we let it "swim" up.              
* Method swim:     
    Basically we compare the value with its parent nodes. If smaller we switch. 
   ```java
       private void swim(int k){
        while( k > 1 &&  ( pq[k/2].compareTo(pq[k]) > 0  ) ){
            // k > 1 意味着 新节点到了root就不能更上浮了。
            //如果新的堆节点比自己的父节点小，那么就交换之。
            swap(k/2,k);
            // k 用作指针，追踪之。
            k = k/2;
        }
    }
   ```
* Method delete: 
    We record the first one, and we swap it with the last one.       
    And then we remove the new last one.      
    And we let the new first one sink down.      
* Method sink :   
    We compare the value of its children node, is larger we sink.    
   ```java
       private void sink(int k){
        // k 输入其实就是等于 1;
        while(2*k <= N){
            int j = 2*k; // j 是 k 的 左子节点
            
            //找到一个局部最大的节点
            //为什么这么做有效呢？因为heap 有点贪心的意思
            //heap是半有序，这也是为什么我们可以写比BST少很多的代码
            //第一层有最小
            //次小 一定在第二层 //第三小 一定在第二层或者第三层
            
            // 这个 if 是找个小一点的儿子。
            if( j < N && pq[j].compareTo( pq[j+1] ) > 0 ){
                j++;
            }
            //如果 k 的值 并不比它的儿子大 那么 k 也算找到位置了。
            if( !(pq[k].compareTo(pq[j]) < 0  ){
                break;
            }
            // 否则的话交换 k, j 的值和位置。继续向下比较
            swap(k,j);
            k = j;
        }
    }
   ```
